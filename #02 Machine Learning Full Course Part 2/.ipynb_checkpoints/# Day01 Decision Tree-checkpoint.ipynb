# Decision Tree
- Root Node: It represents teh entire population or sample and this further get divided into two or more homogeneous sets
- Splitting: It is a Process of dividing a node into two ro more sub - nodes
- Decision Node: When a sub-node splits into further sub-nodes
- Leaf/Terminal Node: Nodes do not split
- Pruning: When we remove sub-nodes of a decision node,
- Branch / subtree: A subsection of entire tree
- Parent and Child Node:
- Entropy / gini index

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"E:\subscription_purchase_data.csv")
dataset.head()
```

```python
dataset.shape
```

```python
x = dataset.iloc[:, :-1]
y = dataset['Purchased']
```

```python
# for scalling
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x = pd.DataFrame(sc.fit_transform(x), columns=x.columns)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)*100
```

```python
# 56	72662	1
# 1	46	118506	1
# 2	32	32688	0

dt.predict([[43	,69726]])
```

```python
dataset['Purchased'].value_counts()
```

```python
dataset.tail()
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(50, 50))
plot_tree(dt)
plt.savefig('Decision Tree.png')
plt.show()
```

```python
dt1 = DecisionTreeClassifier(criterion='entropy')
dt1.fit(x_train, y_train)
```

```python
dt1.score(x_test, y_test)*100
```

```python
sns.scatterplot(x='Age', y='Salary', data=dataset, hue='Purchased')
plt.show()
```

```python
# to check non linear line
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=dt1)
plt.show()
```

# Decision Tree Pre/post Pruning 

```python
dt.score(x_train, y_train)*100
```

```python
dt.score(x_test, y_test)*100
```

this means model is over fitted

```python
# Lets apply pre prunning first
Dt = DecisionTreeClassifier(max_depth=5)
Dt.fit(x_train, y_train)
```

```python
Dt.score(x_train, y_train)*100
```

```python
Dt.score(x_test, y_test)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=Dt)
plt.show()
```

```python
plot_tree(Dt)
plt.show()
```

```python
# now lets check the various accuracies to keep the one which is highest
for i in range(1, 20):
    DT = DecisionTreeClassifier(max_depth=i)
    DT.fit(x_train, y_train)
    print(DT.score(x_train, y_train)*100, DT.score(x_test, y_test)*100, i) # also check overfiting
```

# Decion Tree(Regression)

```python
dataset = pd.read_csv("E:\\experience.csv")
dataset.head()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset['salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
dt.score(x_train, y_train)*100
```

```python
dt.score(x_test, y_test)*100
```

```python
plot_tree(dt)
plt.show()
```
