> Since transforming the data through pandas gave us bolean formate that requires some further steps to convert it to binary formate but we can do this while simply using scikitlearn library that will do it in few steps

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ohe.fit_transform(en_data)
```

> Now this gave me sparse matrix that means data is mostly in the formate of 0 and 1 now lets convert this into an array

```python
DF = ohe.fit_transform(en_data).toarray()
pd.DataFrame(DF, columns=['sex_male', 'sex_female', 'S', 'C', 'Q'])
```

> now in order to delete the column first we need following things

```python
ohe = OneHotEncoder(drop='first')
D = ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(D, columns = ['male', 'C', 'Q'])
```

## **Ordinal Encoding**

```python
df = pd.DataFrame({"Size": ['s', 'm', 'l', 'xl', 'xxl', 's', 'l', 'm', 'xxl', 'l', 'xl', 'm']})
df
```

```python
ord_data = [['s', 'm', 'l', 'xl', 'xxl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
df['Size-encoded']=oe.fit_transform(df)
```

```python
df
```

```python
ord_data1 = {"s":0, "m":1, "l":2, "xl":3, "xxl":4}
df['Size_enc_map']=df['Size'].map(ord_data1)
```

```python
df
```
