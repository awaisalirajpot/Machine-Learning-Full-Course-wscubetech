# Multiclass Classification for Logistic Regression
Ovr one vs rest

```python
dataset = pd.read_csv(r"E:\IRIS.csv")
dataset.head()
```

```python
dataset['species'].unique()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
sns.pairplot(data=dataset, hue='species')
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset['species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## OVR Method

```python
from sklearn.linear_model import LogisticRegression
```

```python
Lr = LogisticRegression(multi_class = 'ovr')
Lr.fit(x_train, y_train)
```

```python
Lr.score(x_test, y_test)
```

## Multinomial

```python
lr1 = LogisticRegression(multi_class = 'multinomial')
lr1.fit(x_train, y_train)
```

```python
lr1.score(x_test, y_test)
```

# Confusion Matrix
Batter Analysis for situations in which model succeded and situations in which model failed

# Confusion Matrix

```python
dataset = pd.read_csv("E:\\nonlinear_binary_classification.csv")
dataset.head()
```

```python
x = dataset.iloc[:, :-1]
y = dataset['Target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
LR = LogisticRegression()
LR.fit(x_train, y_train)
```

```python
LR.score(x_test, y_test)*100
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

```python
cf=confusion_matrix(y_test, LR.predict(x_test))
```

```python
sns.heatmap(cf, annot=True)
```

```python
precision_score(y_test, LR.predict(x_test))*100
```

```python
recall_score(y_test, LR.predict(x_test))*100
```

```python
f1_score(y_test, LR.predict(x_test))**100
```
