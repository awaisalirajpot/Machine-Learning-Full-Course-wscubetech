# Polynomial Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('E:\\salary.csv')
dataset.head()
```

```python
plt.scatter(dataset['Level'], dataset['Salary'])
plt.show()
```

dataset.corr()

```python
dataset.corr()
```

```python
x = dataset[["Level"]]
y = dataset['Salary']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree = 2)
```

```python
pf.fit(x)
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
pred = lr.predict(x)
```

```python
plt.scatter(dataset['Level'], dataset['Salary'])
plt.plot(dataset['Level'], pred, color='red')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

# Cost Function
- A cost function is an important parameter taht determines how well a machne model performs for a given dataset.
- Cost function is a measure of how wrong the model is in estimating the relationship between x(input) and y(output) parameter
**Types**
1. Regression Cost Function
   - MSE (mean square error)
   - RMSE (Root mean square error)
   - MAE (Mean Absolute Error)
   - R sqaure Accuracy
3. Classification Cost Functions
   1. Binary Classification Cost Functions
      - Binary Cross Entropy Cost Function
   3. Multi-class classification cost functions
      - Log loss Function
